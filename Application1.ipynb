```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('data.csv', delimiter=';',encoding = "ISO-8859-1")
```

```python
data
```

```python
fig = px.histogram(
data,
x="duration",
nbins=26, # Nombre de bins
title="Histogramme duration"
)
fig.show()
```

```python
fig = px.histogram(
data,
x="duration",
nbins=26, # Nombre de bins
title="Histogramme duration",histnorm='percent', color="y"
)

fig.show()
```

```python
fig = px.box(
data,
x="duration",
title="Boxplot duration",
color="y"
)
fig.show()
```

```python
fig = px.box(
data,
x="age",
title="Boxplot age",
color="y"
)
fig.show()
```
